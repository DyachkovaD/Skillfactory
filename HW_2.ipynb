```python
from random import randint

class BoardException(Exception):
    pass

class BoardBuzyException(BoardException):
    def __str__(self):
        return f'Клетка занята'

class BoardOutException(BoardException):
    def __str__(self):
        return f"Вы пытаетесь выстрелить за доску!"

class BoardWrongShipException(BoardException):
    pass
```

```python
class Dot:    
    def __init__(self, x, y):
        self.x = x        
        self.y = y  
    
    def __eq__(self, other):        
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Dot({self.x}, {self.y})"

class Ship:
    def __init__(self, edge, l, ort): 
        self.edge = edge # class Dot(x, y)
        self.l = l #int
        self.ort = ort # 0 - horizontal 1 - vertical
        self.lives = l

    @property
    def dots(self):
        ship_dots = []

        for i in range(self.l):
            if self.ort:
                dot = Dot(self.edge.x, self.edge.y + i)
                ship_dots.append(dot)
            else:
                dot = Dot(self.edge.x + i, self.edge.y)
                ship_dots.append(dot)
        return ship_dots

class Board:
    def __init__(self, show=True, size=6):
        self.field = [["~" for i in range(size)] for j in range(size)]
        self.show = show
        self.ships = []
        self.on_ships = 0
        self.buzy = []
        self.count = 0

    def __str__(self):
        f = ''
        f += '   1   2   3   4   5   6 '
        for i, row in enumerate(self.field):
            f  += f"\n{i + 1} | " + " | ".join(row) + " | "

        if not self.show:
            f = f.replace(" ▄ ", "~")
        return f

    def out(self, dot):
        return not ((0 <= dot.x <= 5) or (0 <= dot.y <= 5))
    
    def contour(self, ship, show=False):
        for dot in ship.dots:
            for x in range(dot.x - 1, dot.x + 2):
                for y in range(dot.y - 1, dot.y + 2):
                    if 0 <= x <= 5 and 0 <= y <= 5 and Dot(x, y) not in self.buzy:
                        if show:
                            self.field[x][y] = '•'
                    self.buzy.append(Dot(x, y))

    def add_ship(self, ship):
        for d in ship.dots:
            if self.out(d) or d in self.buzy:
                raise BoardWrongShipException()
        for d in ship.dots:
            self.field[d.x][d.y] = " ▄ "

        self.ships.append(ship)
        self.contour(ship)
        self.on_ships += 1

    def shot(self, dot):
        if self.out(dot):
            raise BoardOutException()
        if dot in self.buzy:
            raise BoardBuzyException()

        self.buzy.append(dot)

        for ship in self.ships:
            if dot in ship.dots:
                ship.lives -= 1
                self.field[dot.x][dot.y] = 'x'

                if ship.lives == 0:
                    self.on_ships -= 1
                    self.contour(ship, show=True)
                    print('Вы потопили корабль')
                    return False
                else:
                    print('Вы подбили корабль')
                    return True
            else:
                self.buzy.append(dot)
                print('Мимо')
                return True
    def begin(self):
        self.buzy = []

class Player:
    def __init__(self, board, e_board):
        self.board = board
        self.e_board = e_board

    def shot_dot(self):
        raise NotImplementedError()

    def move(self):
        while True:
            try:
                dot = self.shot_dot()
                turn = self.e_board.shot(dot)
                return turn
            except BoardException as e:
                print(e)

class User(Player):
    def shot_dot(self):
        while True:
            dot = input('Ваш ход --> ')
            if len(dot.split()) != 2:
                print('Неверный ввод. Введите 2 координаты')
                continue
            x, y  = dot.split()
            if not (x.isdigit() or y.isdigit()):
                print('Неверный ввод. Введите 2 числа в формате: х у')
                continue
            return Dot(int(x) - 1, int(y) - 1)
        
class AI(Player):
    def shot_dot(self):
        dot = Dot(randint(0, 5), randint(0, 5))
        print(f'Ход соперника: {dot.x + 1} {dot.y + 1}')
        return dot
        
class Game:
    def __init__(self, size=6):
        self.size = size
        user_board = self.random_board()
        ai_board = self.random_board()
        ai_board.show = False

        self.user = User(user_board, ai_board)
        self.ai = AI(ai_board, user_board)

    def try_board(self):
        lens = [3, 2, 2, 1, 1, 1, 1]
        board = Board(size=self.size)
        attampts = 0
        for l in lens:
            while True:
                attampts += 1
                if attampts > 2000:
                    return None
                ship = Ship(Dot(randint(0, self.size), randint(0, self.size)), l, randint(0, 1))
                try:
                    board.add_ship(ship)
                    break
                except BoardWrongShipException:
                    pass
        board.begin()
        return board

    def random_board(self):
        board = None
        while board is None:
            board = self.try_board()
        return board
    
    def greet(self):
        print(f'''
                ~~~~~~~~~~~~~~~~~~~
                  ◄ M••рской б•й ►
                ~~~~~~~~~~~~~~~~~~~

                формат хода: х у
                x - номер столбца
                y - номер строки
                ''')

    def play(self):
        turn = 0
        while True:
            print('Твоё поле: ')
            print(self.user.board)
            print('Поле противника: ')
            print(self.ai.board)
            if not turn & 1:
                print('Твой ход --> ')
                repeat = self.user.move()
            else:
                print('Ход противника --> ')
                repeat = self.ai.move()

            if repeat:
                turn -= 1

            if user_board.on_ships == 0:
                print('Вы проиграли!')
                break

            if ai_board.on_ships == 0:
                print('Вы выиграли!')
                break
                
    def start(self):
        self.greet()
        self.play()
             
```
