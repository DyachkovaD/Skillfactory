```python
class Dot:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.figure = "-"
    
    def __eq__(self, x, y):
        return self.x == x and self.y == y

class Ship:
    def __init__(self, lenght, firstDot, direction): #firstDot это tuple вида(2, 6); direction это string вида 'vertical'/'horizontal'
        self.lenght = lenght
        self.firstDot = firstDot
        self.direction = direction
        self.lives = lenght
        self.dots = self.dot()
        
    def dot(self):
        self.dots = []
        if self.lenght == 1:
            self.firstDot = Dot(*self.firstDot)
            self.firstDot.figure = "◘"
            self.dots.append(self.firstDot)
            return self.dots
        self.dots.append(self.firstDot)
        for i in range(self.lenght - 1):
            if self.direction == "vertical":
                self.dots.append((self.firstDot[0], self.firstDot[1] + i + 1))
            else:
                self.dots.append((self.firstDot[0] + i + 1, self.firstDot[1]))
        for i in range(len(self.dots)):
            self.dots[i] = Dot(*self.dots[i])
            self.dots[i].figure = "◘"
        return self.dots
        
class Board:
    def __init__(self):
        self.list_of_dots = [[Dot(x, y) for x in range(1, 7)] for y in range(1, 7)]
        self.list_of_ships = []
        self.hid = True
        self.ships_on_board = 0
        
    def add_ship(self, ship):
        for ship_dot in ship.dots:
            self.list_of_dots[ship_dot.y - 1][ship_dot.x - 1].figure = "◘"
        self.list_of_ships.append(ship)
        self.ships_on_board += 1
        
        
    def show_bord(self):
        for i in self.list_of_dots:
            for j in i:
                print(j.figure, end=" ")
            print()
```

```python
s = Ship(3, (1, 2), "horizontal")
b = Board()
b.add_ship(s)
b.show_bord()
```
